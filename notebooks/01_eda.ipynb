```python
def add_engineered_features(data):
    data = data.copy()

    # Create a cardiometabolic risk score
    data["CardioMetabolicRisk"] = (
        data["HighBP"] +
        data["HighChol"] +
        data["HeartDiseaseorAttack"] +
        data["Stroke"]
    )

    # Create a lifestyle risk score
    data["LifestyleRisk"] = (
        data["Smoker"] +
        (1 - data["PhysActivity"]) +
        (1 - data["Fruits"]) +
        (1 - data["Veggies"]) +
        data["HvyAlcoholConsump"]
    )

    # Create a healthcare access risk score
    data["HealthcareAccessRisk"] = (
        (1 - data["AnyHealthcare"]) +
        data["NoDocbcCost"]
    )

    # Create an overall health burden score
    data["HealthBurden"] = (
        data["GenHlth"] +
        data["MentHlth"] +
        data["PhysHlth"] +
        data["DiffWalk"]
    )

    # Create a socioeconomic status score
    data["SocioEconomicScore"] = (
        data["Education"] +
        data["Income"]
    )

    # Create BMI categories using clinical BMI thresholds
    data["BMI_Category"] = pd.cut(
        data["BMI"],
        bins=[0, 18.5, 25, 30, 100],
        labels=[0, 1, 2, 3]
    ).astype(int)

    return data
```

## Pipeline and Outlier Setup

Feature engineering is applied through `FunctionTransformer`. Scaling is included only for Logistic Regression and Linear SVC, while Random Forest receives the engineered numeric features directly.

Outlier removal is applied inside each training fold before undersampling. Validation folds are not filtered, so evaluation still reflects real unseen data.


```python
feature_engineering_step = FunctionTransformer(add_engineered_features, validate=False)

logistic_pipeline = Pipeline([
    ("feature_engineering", feature_engineering_step),
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        class_weight="balanced",
        max_iter=3000,
        random_state=42
    ))
])

random_forest_pipeline = Pipeline([
    ("feature_engineering", feature_engineering_step),
    ("model", RandomForestClassifier(
        n_estimators=300,
        min_samples_leaf=5,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ))
])

svc_pipeline = Pipeline([
    ("feature_engineering", feature_engineering_step),
    ("scaler", StandardScaler()),
    ("model", LinearSVC(
        C=0.5,
        class_weight="balanced",
        dual=False,
        max_iter=5000,
        random_state=42
    ))
])

base_pipelines = {
    "Logistic Regression": logistic_pipeline,
    "Random Forest": random_forest_pipeline,
    "Linear SVC": svc_pipeline
}
```

## Undersampling Ensemble Function

For each fold, every sub-model trains on all minority rows and one rotating chunk of the majority class. The number of sub-models is calculated from the fold's class balance, with a minimum of 10.


```python
def remove_continuous_outliers_iqr(X, y, continuous_columns, iqr_multiplier=1.5):
    """Remove training rows with IQR outliers in continuous columns."""
    X_filtered = X.copy()
    y_filtered = pd.Series(y).copy()
    keep_mask = pd.Series(True, index=X_filtered.index)

    for column in continuous_columns:
        q1 = X_filtered[column].quantile(0.25)
        q3 = X_filtered[column].quantile(0.75)
        iqr = q3 - q1

        if iqr == 0:
            continue

        lower_bound = q1 - iqr_multiplier * iqr
        upper_bound = q3 + iqr_multiplier * iqr
        keep_mask &= X_filtered[column].between(lower_bound, upper_bound)

    removed_rows = int((~keep_mask).sum())

    return (
        X_filtered.loc[keep_mask].reset_index(drop=True),
        y_filtered.loc[keep_mask].reset_index(drop=True),
        removed_rows
    )


# BMI is the only truly continuous column in this dataset.
continuous_columns = ["BMI"]


def calculate_n_classifiers(y, min_classifiers=10):
    class_counts = pd.Series(y).value_counts()
    majority_class = class_counts.idxmax()
    majority_count = class_counts.max()
    largest_minority_count = class_counts.drop(index=majority_class).max()
    return max(min_classifiers, math.ceil(majority_count / largest_minority_count))


def set_pipeline_random_state(estimator, random_state):
    params = estimator.get_params()
    random_state_params = {
        name: random_state
        for name in params
        if name.endswith("random_state")
    }
    if random_state_params:
        estimator.set_params(**random_state_params)
    return estimator


def get_positive_scores(fitted_model, X):
    if hasattr(fitted_model, "predict_proba"):
        return fitted_model.predict_proba(X)[:, 1]
    return fitted_model.decision_function(X)


def summarize_binary_metrics(y_true, y_pred):
    report = classification_report(
        y_true,
        y_pred,
        output_dict=True,
        zero_division=0
    )

    return {
        "negative_precision": report["0"]["precision"],
        "negative_recall": report["0"]["recall"],
        "positive_precision": report["1"]["precision"],
        "positive_recall": report["1"]["recall"],
        "positive_f1": report["1"]["f1-score"],
        "macro_f1": f1_score(y_true, y_pred, average="macro"),
        "balanced_accuracy": balanced_accuracy_score(y_true, y_pred)
    }


def evaluate_undersampling_ensemble(
    model_name,
    base_pipeline,
    X,
    y,
    cv,
    min_classifiers=10,
    random_state=42,
    outlier_columns=None,
    iqr_multiplier=1.5
):
    y = pd.Series(y).reset_index(drop=True)
    oof_pred = np.zeros(len(y), dtype=int)
    oof_scores = np.zeros(len(y), dtype=float)
    fold_model_counts = []
    fold_outliers_removed = []

    for fold_number, (train_idx, val_idx) in enumerate(cv.split(X, y), start=1):
        X_train_fold = X.iloc[train_idx].reset_index(drop=True)
        X_val_fold = X.iloc[val_idx]
        y_train_fold = y.iloc[train_idx].reset_index(drop=True)

        if outlier_columns:
            X_train_fold, y_train_fold, removed_rows = remove_continuous_outliers_iqr(
                X_train_fold,
                y_train_fold,
                continuous_columns=outlier_columns,
                iqr_multiplier=iqr_multiplier
            )
        else:
            removed_rows = 0

        fold_outliers_removed.append(removed_rows)

        majority_class = y_train_fold.value_counts().idxmax()
        majority_indices = y_train_fold[y_train_fold == majority_class].index.to_numpy(copy=True)
        minority_indices = y_train_fold[y_train_fold != majority_class].index.to_numpy(copy=True)

        n_classifiers = calculate_n_classifiers(y_train_fold, min_classifiers=min_classifiers)
        fold_model_counts.append(n_classifiers)

        rng = np.random.default_rng(random_state + fold_number)
        rng.shuffle(majority_indices)
        majority_chunks = np.array_split(majority_indices, n_classifiers)

        fold_score_sum = np.zeros(len(X_val_fold), dtype=float)

        for model_number, majority_chunk in enumerate(majority_chunks):
            sampled_indices = np.concatenate([minority_indices, majority_chunk])
            rng.shuffle(sampled_indices)

            model = clone(base_pipeline)
            model = set_pipeline_random_state(model, random_state + fold_number + model_number)
            model.fit(X_train_fold.iloc[sampled_indices], y_train_fold.iloc[sampled_indices])

            fold_score_sum += get_positive_scores(model, X_val_fold)

        average_scores = fold_score_sum / n_classifiers
        oof_scores[val_idx] = average_scores

        if average_scores.min() >= 0 and average_scores.max() <= 1:
            oof_pred[val_idx] = (average_scores >= 0.5).astype(int)
        else:
            oof_pred[val_idx] = (average_scores >= 0).astype(int)

    metrics = summarize_binary_metrics(y, oof_pred)

    print("=" * 70)
    print(model_name)
    print("Models per fold:", fold_model_counts)
    print("Outliers removed per fold:", fold_outliers_removed)
    print(classification_report(y, oof_pred, zero_division=0))
    print("Confusion Matrix:")
    print(confusion_matrix(y, oof_pred))
    print("Macro F1:", metrics["macro_f1"])
    print("Balanced Accuracy:", metrics["balanced_accuracy"])

    return {
        "model": model_name,
        "predictions": oof_pred,
        "scores": oof_scores,
        "models_per_fold": fold_model_counts,
        "outliers_removed_per_fold": fold_outliers_removed,
        **metrics
    }
```

## Threshold Selection

The default threshold may not be best for screening. This step evaluates multiple thresholds for each model and selects the best threshold by macro F1. If the goal is maximum screening recall, the threshold can be chosen from the table using `positive_recall` instead.


```python
threshold_results = []

for model_name, output in ensemble_outputs.items():
    scores = output["scores"]

    if scores.min() >= 0 and scores.max() <= 1:
        thresholds = np.arange(0.10, 0.91, 0.05)
    else:
        thresholds = np.quantile(scores, np.arange(0.10, 0.91, 0.05))

    for threshold in thresholds:
        y_pred_threshold = (scores >= threshold).astype(int)
        metrics = summarize_binary_metrics(y_train_binary, y_pred_threshold)

        threshold_results.append({
            "model": model_name,
            "threshold": round(float(threshold), 4),
            **metrics
        })

threshold_results_df = pd.DataFrame(threshold_results)
threshold_results_df.sort_values("macro_f1", ascending=False).head(15)
```

# Final Test Evaluation

Based on cross-validation and threshold tuning, the selected model is:

- Model: Random Forest undersampling ensemble
- Decision threshold: `0.6000`

The final model is trained on the full training set, then evaluated once on the held-out test set.


```python
selected_model_name = "Random Forest"
selected_threshold = 0.60


def fit_undersampling_ensemble_for_deployment(
    base_pipeline,
    X,
    y,
    min_classifiers=10,
    random_state=42,
    outlier_columns=None,
    iqr_multiplier=1.5
):
    y = pd.Series(y).reset_index(drop=True)
    X_train_final = X.reset_index(drop=True)

    if outlier_columns:
        X_train_final, y_train_final, removed_rows = remove_continuous_outliers_iqr(
            X_train_final,
            y,
            continuous_columns=outlier_columns,
            iqr_multiplier=iqr_multiplier
        )
    else:
        y_train_final = y
        removed_rows = 0

    majority_class = y_train_final.value_counts().idxmax()
    majority_indices = y_train_final[y_train_final == majority_class].index.to_numpy(copy=True)
    minority_indices = y_train_final[y_train_final != majority_class].index.to_numpy(copy=True)

    n_classifiers = calculate_n_classifiers(y_train_final, min_classifiers=min_classifiers)

    rng = np.random.default_rng(random_state)
    rng.shuffle(majority_indices)
    majority_chunks = np.array_split(majority_indices, n_classifiers)

    estimators = []

    for model_number, majority_chunk in enumerate(majority_chunks):
        sampled_indices = np.concatenate([minority_indices, majority_chunk])
        rng.shuffle(sampled_indices)

        model = clone(base_pipeline)
        model = set_pipeline_random_state(model, random_state + model_number)
        model.fit(X_train_final.iloc[sampled_indices], y_train_final.iloc[sampled_indices])
        estimators.append(model)

    return {
        "model_name": selected_model_name,
        "threshold": selected_threshold,
        "estimators": estimators,
        "n_classifiers": n_classifiers,
        "outliers_removed": removed_rows,
        "outlier_columns": outlier_columns,
        "iqr_multiplier": iqr_multiplier,
        "class_mapping": {
            0: "No diabetes",
            1: "Prediabetes or diabetes"
        }
    }


def score_undersampling_ensemble(ensemble, X):
    score_sum = np.zeros(len(X), dtype=float)

    for estimator in ensemble["estimators"]:
        score_sum += get_positive_scores(estimator, X)

    return score_sum / len(ensemble["estimators"])


def predict_undersampling_ensemble(ensemble, X):
    scores = score_undersampling_ensemble(ensemble, X)
    predictions = (scores >= ensemble["threshold"]).astype(int)
    return predictions, scores
```

```python
final_rf_ensemble = fit_undersampling_ensemble_for_deployment(
    base_pipeline=random_forest_pipeline,
    X=X_train_raw,
    y=y_train_binary,
    min_classifiers=10,
    random_state=642,
    outlier_columns=continuous_columns,
    iqr_multiplier=1.5
)

print("Final model:", final_rf_ensemble["model_name"])
print("Decision threshold:", final_rf_ensemble["threshold"])
print("Number of ensemble classifiers:", final_rf_ensemble["n_classifiers"])
print("Training outliers removed:", final_rf_ensemble["outliers_removed"])
```

```python
final_test_pred, final_test_scores = predict_undersampling_ensemble(
    final_rf_ensemble,
    X_test_raw
)

final_test_metrics = summarize_binary_metrics(y_test_binary, final_test_pred)

print("Final Random Forest Undersampling Ensemble: Test Set")
print(classification_report(y_test_binary, final_test_pred, zero_division=0))
print("Confusion Matrix:")
print(confusion_matrix(y_test_binary, final_test_pred))

final_test_results_df = pd.DataFrame([{
    "model": selected_model_name,
    "threshold": selected_threshold,
    **final_test_metrics
}])

final_test_results_df
```
